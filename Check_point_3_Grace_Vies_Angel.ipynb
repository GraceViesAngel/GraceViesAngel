```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/drive/MyDrive/AVD/student_spending (1) (1).csv")

df
```

# Construct Data

```python
df['total_funds'] = df['monthly_income'] + df['financial_aid']
print(df.head())
```

# Data Reduction

```python
df = df.drop('gender', axis=1)
```

```python
df
```

# Visualisasi Data

```python
# Histogram - Distribusi usia mahasiswa
plt.figure(figsize=(8, 5))
sns.histplot(df['age'], bins=20, kde=True, color='blue')
plt.xlabel("Usia Mahasiswa")
plt.ylabel("Jumlah Mahasiswa")
plt.title("Distribusi Usia Mahasiswa")
plt.show()
```

Insight:
1. Mayoritas mahasiswa berada dalam retang usia 18-24 tahun, yang merupakan usia tipikal untuk jenjang pendidikan tinggi
2. Jika ada outlier (usia jauh lebih tua atau lebih muda), ini bisa disebabkan oleh mahasiswa yang mengambil kuliah lebih awal atau terlambat.

tindakan yang diambil:
1. kampus dapat menyesuaikan program finansial dan bantuan keuangan berdasarkan rentang usia yang paling dominan.
2. Mahasiswa yang lebih tua atau lebih muda dari mayoritas mungkin membutuhkan dukungan atau kebijakan khusus dalam pengelolaan keuangan.

```python
# Bar Chart - Distribusi Pengeluaran Berdasarkan Kategori
plt.figure(figsize=(10, 5))
df[['tuition', 'housing', 'food', 'transportation', 'books_supplies',
    'entertainment', 'personal_care', 'technology', 'health_wellness', 'miscellaneous']].mean().sort_values().plot(kind='bar', color='skyblue')
plt.xlabel("Kategori Pengeluaran")
plt.ylabel("Rata-rata Pengeluaran")
plt.title("Distribusi Pengeluaran Berdasarkan Kategori")
plt.xticks(rotation=45)
plt.show()
```

Insight:
1. Pengeluaran terbesar mahasiswa berada di biaya kuliah dan tempat tinggal
2. Pengeluaran untuk hiburan dan transportasi lebih kecil dibanding kebutuhan poko

Tindakan yang diambil:
1. Mahasiswa harus mengidentifikasi kategori pengeluaran yang paling besar dan menyesuaikan anggaran agar lebih efisien dengan memanajemen keuangan yang lebih efisien
2. Kampus dapat memberikan edukasi keuangan tentang cara mengelola keuangan dan menghemat biaya hidup.

```python
# Boxplot - Perbandingan Pengeluaran Berdasarkan Tahun Studi
plt.figure(figsize=(10, 6))
sns.boxplot(x='year_in_school', y='total_funds', hue='year_in_school', data=df, palette="pastel")
plt.xlabel("Tahun Studi")
plt.ylabel("Total Dana (Income + Financial Aid)")
plt.title("Perbandingan Total Dana Berdasarkan Tahun Studi")
plt.xticks(rotation=45)
plt.show()
```

Insight:
1. Tidak ada perbedaan signifikan dalam total dana (pendapatan + bantuan keuangan) di antara tahun studi
2. Beberapa mahasiwa dalam kelompok tertentu mungkin memiliki dana lebih tinggi atau lebih rendah dibanding lainnya

Tindakan yang diambil:
1. Kalau misalkan ada mahasiswa tingkat awal memiliki dana lebih rendah, pihak kampus dapat mempertimbangkan kebijakan subsidi bagi mahasiswa baru
2. Mahasiswa tingkat lanjut bisa memanfaatkan peluang beasiswa atau pekerjaan paruh waktu untuk meningkatkan pendapatan mereka

```python
# Pie Chart - Proporsi mahasiswa berdasarkan tahun studi
plt.figure(figsize=(6, 6))
df['year_in_school'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['lightblue', 'lightgreen', 'coral', 'gold'])
plt.title("Proporsi Mahasiswa berdasarkan Tahun Studi")
plt.ylabel("")
plt.show()
```

Insight:
1. jumlah mahasiswa mungkin tidak merata di tiap tahun studi jika ada ketimpangan, ini bisa jadi karena tingkat dropout yang tinggi atau seleksi akademik

Tindakan yang diambil:
1. kampus dapat mengevaluasi kebijakan akademik atau keuangan untuk mengurangi dropout
2. Bantuan keuangan bisa disesuaikan untuk mahasiswa tingkat akhir yang mungkin lebih kesulitan secara finansial

```python
# Scatter Plot - Hubungan antara pendapatan dan total pengeluaran
plt.figure(figsize=(8, 5))
sns.scatterplot(x='monthly_income', y='total_funds', data=df, alpha=0.6, color='red')
plt.xlabel("Pendapatan Bulanan")
plt.ylabel("Total Dana")
plt.title("Hubungan antara Pendapatan dan Total Dana")
plt.show()
```

Insight:
1. beberapa mahasiswa dengan pendapatan rendah memiliki total dana tinggi karena mendapatkan bantuan keuangan yang besar
2. Jika tidak ada hubungan jelas, berarti pengeluaran mahasiswa dipengaruhi faktor lain seperti gaya hidup atau kebiasaan konsumsi

Tindakan yang diambil:
1. Mahasiswa perlu mengontrol pengeluaran, terutama jika mengalami kenaikan pendapatan dari pekerjaan sampingan atau beasiswa
2. Pihak universitas dapat meningkatkan program beasiswa berbasis kebutuhan agar distribusi bantuan lebih tetap sasaran

```python
# Heatmap - Korelasi antara pendapatan, bantuan keuangan, dan total pengeluaran
plt.figure(figsize=(8, 6))
sns.heatmap(df[['monthly_income', 'financial_aid', 'total_funds']].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Korelasi antara Pendapatan, Bantuan Keuangan, dan Total Dana")
plt.show()
```

Insight:
1. Korelasi antara pendapatan dan bantuan keuangan rendah, menunjukkan bahwa bantuan keuangan tidak selalu diberikan kepada mereka yang memiliki pendapatan rendah
2. Total dan lebih dipengaruhi oleh bantuan keuangan dibanding pendapatan pribadi

Tindakan yang diambil:
1. Universitas atau pemerintah bisa mengevaluasi kebijakan pemberian bantuan keuangan agar lebih tepat sasaran ke mahasiswa yang benar-benar membutuhkan.
2. Mahasiswa yang belum mendapatkan bantuan keuangan sebaiknya mencari program beasiswa atau skema dukungan lain untuk meningkatkan stabilitas finansial.
