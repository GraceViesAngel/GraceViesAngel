# statistik deskriptif dan analisis korelasi

```python
print('menampilkan statistik deskriptif dari dataset:')
print(df.describe())
```

```python
kolom_unik = 'food'
if kolom_unik in df.columns:
  print(f'\nJumlah nilai unik dalam kolom {kolom_unik}:')
  print(df[kolom_unik].value_counts())
else:
  print(f'\nKolom {kolom_unik} tidak ditemukan dalam dataset.')

print('\nmenghitung rata-rata dari setiap kolom:')
print(df.mean(numeric_only=True))

print('\nmenghitung median dari setiap kolom:')
print(df.median(numeric_only=True))

print('\nmenghitung standar deviasi dari setiap kolom:')
print(df.std(numeric_only=True))

print('\nmenghitung korelasi antara setiap pasangan kolom:')
print(df.corr(numeric_only=True))
```
